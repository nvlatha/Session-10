```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy 
import seaborn as sns
```

```python
# Read data from the url and print
url= "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
```

```python
#1. Create a pie chart presenting the male/female proportion
# use group by to get the total sum under each categoty
# Colors we can define , Title will have the title of the chart, Startanagle = will define thestart angle for the chart
# Lambda is used to get the percentage in both genders
gender = titanic.groupby('sex').size()
sns.set()
colors = ['pink', 'blue']
gender.plot(kind='pie', title='Male & Female proportion', colors=colors, startangle=90,
         autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*gender.sum()))
plt.show()
```

```python
#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
# Drop the NaN values in Sex, mapping will help to manage colors, xlable, ylable for the X and Y axis 
#title for the title of the chart
titanic = titanic.dropna(subset=['sex'])

mapping = {'male' : 'orange', 'female' : 'green'}
plt.xlabel('Age')
plt.ylabel('Fare paid')
plt.title('Scatterplot with the Fare paid and Age')
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].map(mapping))
plt.show()
```
